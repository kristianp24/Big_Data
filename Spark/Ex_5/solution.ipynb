```python
from pyspark.sql import SparkSession
dataPath = "persons.csv"

session = SparkSession.builder.getOrCreate()

data = session.read.load(dataPath, format="csv", header=True, inferSchema = True)
data.show(10)
#DF solution
selected_data = data.groupby("name").agg({"age":"avg", "name":"count"})
selected_data.show(10)

#SQL solution
data.createOrReplaceTempView("people")
selected_data2 = session.sql("select name, avg(age) as AverageAge, count(name) as Count from people group by name")
selected_data2.show(10)
```
