```python
# Task 1
def function_split(line):
    return line.split("\t")[0].startswith('ho')

if __name__ == "__main__":
    conf= SparkConf().setAppName("Filter for lines")
    sc= SparkContext.getOrCreate(conf=conf) 

    input_file = input_path

    rdd = sc.textFile(input_file)

    parsed_lines = rdd.filter(function_split)
    print(parsed_lines.take(1)[0].split("\t")[0])
    print(f"Number of selected lines {parsed_lines.count()}")

    most_frequent = parsed_lines.map(lambda line: line.split('\t')[1]).max()
    print(f"The maxim freq: {most_frequent}")
```
